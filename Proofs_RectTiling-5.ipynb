```python
from definitions import *
num_cores = multiprocessing.cpu_count()
```

```python
Graphs5 = load_plantri(filename='plantri/10pm4c3')
print(f'Loaded {len(Graphs5)} grahps from plantri.')

inputs = tqdm(Graphs5)
if __name__ == "__main__":
    processed_list = Parallel(n_jobs=num_cores)(delayed(add_distinguished)(G) for G in inputs)
Graphs5 = [G for DGraph in processed_list for G in DGraph]
print(f'Constructed {len(Graphs5)} graphs with a distinguished vertex.')

inputs = tqdm(Graphs5)
if __name__ == "__main__":
    processed_list = Parallel(n_jobs=num_cores)(delayed(first_filter)(G) for G in inputs)
Graphs5 = [Graphs5[i] for i in range(len(Graphs5)) if processed_list[i]]
print(f'There are {len(Graphs5)} quadrilateral graphs left after first filter.')
save_data(Graphs5,'grahps/rec_filtered_5_4.txt')

inputs = tqdm(Graphs5)
if __name__ == "__main__":
    processed_list = Parallel(n_jobs=num_cores)(delayed(has_min_deg)(G,5) for G in inputs)
Graphs5_5 = [Graphs5[i] for i in range(len(Graphs5)) if processed_list[i]]
print(f'There are {len(Graphs5_5)} pentagonal graphs.')
```
